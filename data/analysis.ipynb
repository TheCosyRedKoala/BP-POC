```python
# Importing the necessary packages
import numpy as np                                  # "Scientific computing"
import scipy.stats as stats                         # Statistical tests

import pandas as pd                                 # Data Frame
import vaex

import matplotlib.pyplot as plt                     # Basic visualisation
import seaborn as sns                               # Advanced data visualisation
import altair as alt    
```

```python
# Importing Data
data = pd.read_csv("datareading.csv", delimiter=",")
```

```python
print(data)
```

```python
# Mean
mean = data.TimeDifferenceInMilliseconds.mean()
# Median
median = data.TimeDifferenceInMilliseconds.median()

# Standard Variance
standardVariance = data.TimeDifferenceInMilliseconds.std()

# Quantiles
quantiles = [0.25, 0.5, 0.75]

q1 = data.TimeDifferenceInMilliseconds.quantile(quantiles[0])
q2 = data.TimeDifferenceInMilliseconds.quantile(quantiles[1])
q3 = data.TimeDifferenceInMilliseconds.quantile(quantiles[2])

# Interquantile distance
interQuantileDistance = data.TimeDifferenceInMilliseconds.quantile(quantiles[2]) - data.TimeDifferenceInMilliseconds.quantile(quantiles[0])

# Reporting
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Variance: {standardVariance}")
print(f"Q1: {q1}")
print(f"Q2: {q2}")
print(f"Q3: {q3}")
print(f"Interquantile Distance: {interQuantileDistance}")
```

```python
# Boxplot
sns.boxplot(data=data, x="TimeDifferenceInMilliseconds")
```

```python
# Histogram
sns.displot(x=data.TimeDifferenceInMilliseconds) # Met dichtheidslijn
```

```python
# Dichtheidsdiagram
sns.kdeplot(x=data.TimeDifferenceInMilliseconds)
```

```python
# Stap 1 #
p = 0.95 # Percentage
s = 1.5 # Standaard deviatie van de populatie
m = mean # Sample gemiddelde
n = len(data) # Aantal samples

# Alpha berekenen #
alpha = - (p - 1) # Significatie level

# Degrees of Freedom berekenen #
df = n - 1

# Stap 2 #
t = stats.t.isf(alpha / 2, df)

# Stap 3 #
lowerlimit = m - t * s / np.sqrt(n)
upperlimit = m + t * s / np.sqrt(n) 

print(f"Lower limit: {lowerlimit}")
print(f"Upper limit: {upperlimit}")
```
